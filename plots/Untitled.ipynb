```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Flows

```python
df = pd.read_csv('t1results.csv')
df.head()
```

```python
flows = df[['k', ' flow']].groupby('k').mean()[' flow']
paths = df[['k', ' paths']].groupby('k').mean()[' paths']
time = df[['k', ' time']].groupby('k').mean()[' time']
```

```python
scale = [2**(k+1) for k in range(16)]
plt.plot(scale, flows)
plt.title('calculated flow')
plt.xlabel('graph size')
plt.show()
plt.plot(scale, paths)
plt.title('number of calculated paths')
plt.xlabel('graph size')
plt.show()
plt.plot(scale, time)
plt.title('execution time')
plt.xlabel('graph size')
plt.show()
```

# Matches

```python
df = pd.read_csv('t2results.csv')
df.head()
```

```python
for k in range(3,11):
    kdf = df[df['k'] == k]
    kdf[[' i', ' matches']].groupby(' i').mean().plot(title='k={}'.format(k))
    
```

```python
for i in range(1,11):
    cur_df = df[df[' i'] == i]
    cur_df[['k', ' time']].groupby('k').mean().rename(columns = {' time':'time (ns)'}).plot(title='i={}'.format(i), kind='area')
```
